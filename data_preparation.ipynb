# Data Preparations

```python
import numpy as np
import pandas as pd
```

## Prepare Salary Data

```python
orig_salary = pd.read_csv("zpl_2023.csv", index_col="Год")
orig_salary
```

```python
# sort values
orig_salary = orig_salary.sort_values("2023")
orig_salary
```

```python
DOMAIN_NAMES = [
    "restaurants_and_hotels",
    "education", 
    "healthcare_and_socials",
    "construction_industry",
    "processing_industry",
    "overall",  
    "mining",
    "finances_and_insurances",    
]
```

```python
# transponse df
orig_salary = orig_salary.T

# rename colums
orig_salary.columns.name = None
if orig_salary.columns.shape[0] == len(DOMAIN_NAMES):
    orig_salary.columns = DOMAIN_NAMES
else:
    raise IndexError
```

```python
# get years to the column
orig_salary.index.name = "year"
orig_salary = orig_salary.reset_index(drop=False)

orig_salary.tail(5)
```

## Prepare inflation data

$$Real Income = (1 - Inflation Rate) * Wages$$

```python
inf_data = pd.read_csv("inflation_in_years.csv", sep=";")

# filter data
inf_data = inf_data[2000 <= inf_data.year]
inf_data = inf_data[inf_data.year <= 2023]

# reverce (for convinient)
inf_data = inf_data[::-1]
inf_data = inf_data.reset_index().iloc[:, 1:]
```

```python
inflation_coeffs = 1 - inf_data["inflation"] / 100
inflation_coeffs[3:8]
```

```python
accum = 1
acuum_coeffs = [1, ]
for i in range(1, inf_data.shape[0]):
    accum = accum * inflation_coeffs[i-1]
    acuum_coeffs.append(accum)
```

34234

```python
inf_data["bring_to_2000"] = acuum_coeffs
# inf_data["bring_to_2000"] = 1 / inf_data["bring_to_2000"]
inf_data
```

## Cacl salary with inflation

```python
salary_data = orig_salary.copy()
```

```python
salary_cols = DOMAIN_NAMES
print(f"Cols that affected: \n{salary_cols}")
```

```python
for salary_col in salary_cols:

    # calc salary with inflation
    new_name = salary_col + "_inf"
    salary_data[new_name] = salary_data[salary_col] * inf_data["bring_to_2000"]
    salary_data[new_name] = np.round(salary_data[new_name])
    salary_data[new_name] = salary_data[new_name].astype("int64")

    # bringing to "how it feels in 2023"
    correcting_coeff =  salary_data[salary_col].iloc[-1] / salary_data[salary_col + "_inf"].iloc[-1]

    new_name = new_name + "_adj"
    salary_data[new_name] = salary_data[salary_col + "_inf"] * correcting_coeff
    salary_data[new_name] = np.round(salary_data[new_name])
    salary_data[new_name] = salary_data[new_name].astype("int64")

```

## Add dollar USA
Source: [Динамика официального курса ЦБ](https://cbr.ru/currency_base/dynamics/?UniDbQuery.Posted=True)

```python
dollar_full = pd.read_csv("dollar_full.csv", decimal=",")

dollar_full["data"] = pd.to_datetime(dollar_full["data"], format="%d.%m.%Y")
dollar_full["year_of_measure"] = dollar_full["data"].dt.year

dollar_full.head()
```

```python
dollar_full.dtypes
```

Т.к. измерения курса проводились почти каждый день, можно не заморачиваться и просто посчитать медиану по годам

```python
dollar_years = dollar_full.groupby("year_of_measure").agg({"data":"count", "curs":"median"})
dollar_years.columns = ["cnt_measures", "median_currency_rate"]

dollar_years.index.name = "year"
dollar_years = dollar_years.reset_index(drop=False)

dollar_years
```

```python
print(f"Cols that affected: \n{salary_cols}")

for salary_col in salary_cols:
    new_name = salary_col + "_in_dollars"
    salary_data[new_name] = salary_data[salary_col] / dollar_years["median_currency_rate"]
    salary_data[new_name] = np.round(salary_data[new_name])
    salary_data[new_name] = salary_data[new_name].astype("int64")
```
